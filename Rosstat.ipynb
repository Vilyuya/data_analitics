```python
# сколько раз изменилась  средняя цена на  определенный товар
product = 'Хлеб и булочные изделия из пшеничной муки высшего сорта'
data.loc['2021',product]/ data.loc['1998', product]
```

```python
product = 'Сигареты с фильтром отечественные, за пачку'
data.loc['2021',product]/ data.loc['1998', product]
```

```python
# извлекаем года для временной шкалы  по оси х
years = data.index
years = years[years >= "1998"]
years
```

```python
data.columns
```

```python
# очищение пробелов  в названии 
# лямбда - выражения  (безымянная функция) 

foo = lambda a, b, c: a + b + c
foo(10, 20, 30) 
```

```python
func_list = [
    lambda a: a ** 2, 
    lambda a, b: a * b
]
func_list[0](100), func_list[1](10,20)
```

```python
#  убирает пробел или знак 
s = " python "
s.strip(" ")
```

```python
# удаляем лишние символы пробела из ключей столбцов (названий) 
data.columns = np.array(list(map(lambda key_str: key_str.strip(" "), data.columns)))                    
data.columns                                                
```

```python
def f1(df,  key , idx = '1998'):
    return df.loc[idx:, key]
```

```python
# извлекаем данные по ценам 
col_1 = data.loc["1998":, 'Говядина (кроме бескостного мяса)']
col_2 = f1(data, 'Рыба соленая, маринованная, копченая')
col_3 = f1(data, 'Масло сливочное 5)')
col_4 = f1(data, 'Кирпич красный, за 1000 шт.')
col_5 = f1(data, 'Сыры сычужные твердые и мягкие')
```

```python
import matplotlib.pyplot as plt
```

```python
# визуализация данных


plt.figure(figsize=(12,6))

plt.plot(years, col_1, label ="Говядина")
plt.plot(years, col_2, label ="Рыба")
plt.plot(years, col_3, label ="Масло")
plt.plot(years, col_4, label ="Кирпич")
plt.plot(years, col_5, label ="сыр")

plt.yscale("log")
plt.legend()
plt.show()
```
