## Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

- Простые линейные графики

```python
# 1- мерный массив (вектор-строка)
x_array = np.arange(10)
x_array
```

```python
y = x_array + 10

# визуализация
plt.plot(x_array, y)
plt.show()
```

```python
y = x_array ** 2

# визуализация
plt.plot(x_array, y)
plt.show()
```

```python
y_1 = x_array ** 2
y_2 = x_array + 2
y_3 = x_array / -1
# визуализация
plt.plot(x_array, y_1)
plt.plot(x_array, y_2)
plt.plot(x_array, y_3)
plt.show()
```

```python
# 1-мерный массив из 100 значений
# линейная последовательность в диапозоне от  0  до 10
x = np.linspace(0,10,100)
x
```

```python
x = np.linspace(-10,10,100)
# функция гиперболического тангенса
y_1 = np.tanh(x)

# функция сигмоиды
def sigmoid(x_arr):
    return 1 / (1 + np.exp(-x_arr))

y_2 = sigmoid(x)

# визуализация
plt.plot(x, y_1)
plt.plot(x,y_2)
plt.show()
```

```python
# цвет линий графика
y_1 = np.sin(x)
y_2 = np.cos(x)
y_3 = np.sin(x+1)
y_4 = np.sin(x+2)
y_5 = np.cos(x+3)

plt.plot(x, y_1, color="red") # по названию  цвета на англ 
plt.plot(x, y_2, color="m") # по первой букве цвета из  форматов  rgb cmyk
plt.plot(x, y_3, color="#fa702a") # по 16 - ричной кодировке формата RGB 
plt.plot(x, y_4, color =(0.0, 0.8, 0.7)) # по формату RGB в виде кортежа со значениями от 0 до 1
plt.plot(x, y_5, color="#9933cc") # по 16 - ричной кодировке формата RGB 
plt.show()
```

```python
# стиль линий
y_1 = np.sin(x)
y_2 = np.sin(x+1)
y_3 = np.sin(x+2)
y_4 = np.sin(x+3)


plt.plot(x, y_1, color="red", linestyle='-') # 'solid' сплошная
plt.plot(x, y_2, linestyle='--') # 'dashed' штриховая
plt.plot(x, y_3, linestyle=':') # 'dotted' пунктирная
plt.plot(x, y_4, linestyle='-.' ) # штрих-пунктирная

plt.show()
```

```python
# Объединение настроек цвета и стиля  линий

plt.plot(x, y_1, "g")
plt.plot(x, y_2, "--r") 
plt.plot(x, y_3, ":b") 
plt.plot(x, y_4,"-.y") 

plt.show()
```

```python
# легенда графика 

plt.plot(x, y_1, "g", label="синус")
plt.plot(x, np.cos(x), "--r", label="косинус") 

# диапозон координатной сетки
plt.axis([-15,15, -2, 2])
plt.legend()
plt.show()
```

```python
# подпись оси 

plt.plot(x, y_1, "g", label="синус")
plt.plot(x, np.cos(x), "--r", label="косинус") 

plt.legend()
plt.xlabel("x")
plt.ylabel("sin and cos")
# диапозон координатной сетки
plt.axis([-15,15, -2, 2])


plt.show()
```

```python
# Заголовок графика 

# размеры графика 
plt.figure(figsize=(10,7))

plt.plot(x, y_1, "g", label="синус", marker ="*") #"o"
plt.plot(x, np.cos(x), "--r", label="косинус") 
plt.plot(x, np.cos(x+1), "oy:", label="косинус")
plt.legend()
plt.xlabel("x")
plt.ylabel("sin and cos")
plt.title("Графики синуса и косинуса")

# диапозон координатной сетки
plt.axis([-15,15, -2, 2])


plt.show()
```
