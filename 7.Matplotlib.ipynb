```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

- Диаграмма рассеивания

```python
x = np.linspace(0, 10, 50)
y = np.sin(x)

plt.scatter(x, y)

plt.show()
```

```python
x = np.random.randn(500)
y = np.random.randn(500)

# задаем цвета точкам 
# colors = np.random.random(500)

colors = x

# задаем размеры точек
sizes = np.linspace(0, 50, 500)

plt.scatter(x, y, c=colors, cmap = "Spectral", s= sizes)
plt.show()
```

```python
# обьединение в одном окне графика и диаграммы рассеивания 

x = np.linspace(0, 10 , 1000)
y = np.sin(x) 

# модуляция значений по y
# m = np.random.random(1000)
m = np.random.randn(1000) * 0.5
y1 = y + m

# диаграмма рассеивания 
plt.scatter(x,y1, c= m, cmap="Spectral")

# график
plt.plot(x,y, c="k")



plt.show()
```

```python
# Города Якутии 
pop_dict ={
    "Якутск": 318768,
    "Нерюнгри": 57009,
    "Вилюйск": 11095
}
pop_series = pd.Series(pop_dict)
pop_series
```

```python
# данные по площадям городов
area_series = pd.Series({
    "Якутск": 122,
    "Вилюйск": 15,
    "Нерюнгри": 98.9
       
})
area_series
```

```python
# данные координаты (Lat - широта, Lon - долгота)
lat = {
    "Якутск": 62,
    "Вилюйск": 63,
    "Нерюнгри": 56
}
lon = {
    "Якутск": 129,
    "Вилюйск": 121,
    "Нерюнгри": 124
}

df = pd.DataFrame({"Население": pop_series, "Площадь": area_series, "Широта": lat, "Долгота": lon})
df
```

```python
# Визуализация данных 

# лагарифмирование данных по населению
pop_log = np.log10(df["Население"])

plt.scatter(df["Долгота"], df["Широта"], c=pop_log, s = df["Площадь"])

# цветовая шкала
plt.colorbar()

plt.show()
```
